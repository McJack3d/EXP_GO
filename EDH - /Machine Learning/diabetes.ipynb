```python
import pandas as pd
from sklearn.naive_bayes import GaussianNB
```

```python
df = pd.read_csv("/Users/alexandrebredillot/Documents/GitHub/EXP_GO/EDH - /Machine Learning/diabetes.csv")
```

```python
feature_set = ['pregnant', 'insulin','bmi','age','glucose','bp','pedigree']
features = df[feature_set]
target = df.label
```

```python
from sklearn.model_selection import train_test_split
feature_train,feature_test, target_train, target_test = \
train_test_split(features, target, test_size=0.3)
```

```python
from sklearn.naive_bayes import GaussianNB
# Create a Gaussian Classifier
model = GaussianNB()
# Train the model using the training sets
model.fit(feature_train,target_train)
# Forecast the target variable for given test dataset
predictions = model.predict(feature_test)
# Evaluate the model's accuracy on the training and test sets
train_accuracy = model.score(feature_train, target_train)
print("Accuracy on training set: {:.3f}".format(train_accuracy))
test_accuracy = model.score(feature_test, target_test)
print("Accuracy on test set: {:.3f}".format(test_accuracy))
```
