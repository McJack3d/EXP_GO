```python
import numpy as np 
import pandas as pd
```

```python
x = 2 + 2 * np.random.randn(500)
xo = 10 + 2 * np.random.randn(5)
# concatenate the main data and the outliers so shapes match (result length 505)
y = np.concatenate([x, xo])
df = pd.DataFrame({"a": y})
```

```python
import matplotlib.pyplot as plt
```

```python
plt.plot(df)
plt.show()
```

```python
mad = np.median(np.abs(df - np.median(df)))
print("Median Absolute Deviation:", mad)
```
